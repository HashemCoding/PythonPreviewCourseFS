### Init

```python
import pandas as pd
```

### Get started

```python
df = pd.read_csv('tips.csv')
display(df)
print(df.describe())
```

### Read data

```python
print('sum of tips in 1995 %.2f $' % df['tip'].sum())
```

### Filtern von Daten

```python
filtered_data = df[df['size'] > 2]
display(filtered_data.reset_index(drop=True))
```

```python
selected_data = df.loc[df['smoker'] =="Yes", ['sex']]
selected_data
```

### Functions

```python
df = pd.read_csv('tips.csv')
print('%d rows and %d columns' % df.shape)
print('_____________________________')
display(df.head(10))
display(df.tail(10))
print('_____________________________')
display(df.columns)
display(df.dtypes)
print('_____________________________')
df.info()
```

```python
print ('sum of size: {} people'.format(df['size'].sum()))
print ('highest tip: %.2f $' % df['tip'].max())
print ('lowest tip: %.2f $' % df['tip'].min())
```

### Rename Column

```python
df.rename(columns={'day': 'dayn'}, inplace = True)
df.rename(columns={'dayn': 'day'}, inplace = True)
display(df)
```

### Merge df

```python
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['John', 'Emma', 'Alex', 'Sophia']
})

df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'Age': [30, 25, 35, 28]
})

# Merging the data frames based on the 'ID' column
merged_df = pd.merge(df1, df2, on='ID', how='outer')

print("Merged Data Frame:")
display(merged_df)
```

### Groupby

```python
gdf = df.groupby('sex').aggregate({'sex': 'count'})
display(gdf)
```

### Create Column

```python
df[['bill_2022', 'tip_2022']] = df[['total_bill', 'tip']] * 2.04
display(df)
```

```python
df['percentage'] = df['tip'] / df['total_bill'] * 100
display(df)
```

```python
print ('sum of tips in 2024 %.2f $' % df['tip_2022'].sum())
```

```python
selected_data = df[df['percentage'] >70]
display(selected_data)
```

# Plot

## Matplotlib

```python
import matplotlib as plt
```

```python
pd.options.plotting.backend = "matplotlib"
```

```python
df['rounding'] = (df['total_bill'] + df['tip']).astype('int64') == (df['total_bill'] + df['tip'])
df.groupby('rounding').count()['tip'].plot(kind = 'pie')
```

```python
df['consumption per person'] = df['bill_2022'] / df['size']
df['consumption per person'].plot(kind = 'hist')
```

```python
gdf = df.groupby('sex').aggregate({'sex': 'count'})
gdf.plot(kind = 'bar')
```
