### Functions

```python
df = pd.read_csv('tips.csv')
print('%d rows and %d columns' % df.shape)
print('_____________________________')
display(df.head(10))
display(df.tail(10))
print('_____________________________')
display(list(df.columns))
display(df.dtypes)
print('_____________________________')
df.info()
```

```python
print ('sum of size: {} people'.format(df['size'].sum()))
print ('highest tip: %.2f $' % df['tip'].max())
print ('lowest tip: %.2f $' % df['tip'].min())
```

### Rename Column

```python
#df.rename(columns={'day': 'dayn'}, inplace = True)
df.rename(columns={'dayn': 'day'}, inplace = True)
display(df)
```

### Merge df

```python
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['John', 'Emma', 'Alex', 'Sophia']
})

df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'Age': [30, 25, 35, 28]
})
display(df1)
display(df2)
# Merging the data frames based on the 'ID' column
merged_df = pd.merge(df1, df2, on='ID', how='right')
concat_df = pd.concat([df1, df2], axis=0, ignore_index=True)

print("Merged Data Frame:")
display(merged_df)
display(concat_df)
```

### Groupby

```python
gdf = df.groupby('sex').aggregate({'sex': 'count'})
display(gdf)
```

```python
gdf = df.groupby(['smoker',"sex", "time"]).agg({'total_bill': 'mean', 'tip': 'mean'})
#gdf["neu"] = "sui"
display(gdf)
```

### Create Column

```python
df[['bill_2022', 'tip_2022']] = df[['total_bill', 'tip']] * 2.05
display(df)
```

```python
df['percentage'] = df['tip'] / df['total_bill'] * 100
display(df)
```

```python
print ('sum of tips in 2024 %.2f $' % df['tip_2022'].sum())
```

```python
selected_data = df[df['percentage'] >70]
display(selected_data)
```

# Plot

## Matplotlib

```python
pd.options.plotting.backend = "matplotlib"
```

```python
df['rounding'] = (df['total_bill'] + df['tip']).astype('int64') == (df['total_bill'] + df['tip'])
df.groupby('rounding').count()['tip'].plot(kind = 'pie')
```

```python
df['consumption per person'] = df['bill_2022'] / df['size']
df['consumption per person'].plot(kind = 'pie')
```

```python
gdf = df.groupby('sex').aggregate({'sex': 'count'})
gdf.plot(kind = 'bar')
```

## Plotly

```python
pd.options.plotting.backend = "plotly"
```

```python
df['rounding'] = (df['total_bill'] + df['tip']).astype('int64') == (df['total_bill'] + df['tip'])
df.groupby('rounding').count()['total_bill'].plot(kind = 'bar')
```

```python
df['consumption per person'] = df['bill_2022'] / df['size']
df['consumption per person'].plot(kind = 'hist')
```

```python
gdf=df.groupby('sex').agg({'tip':'sum'}).plot(kind='bar')
display(gdf)
```

### ploting extern

```python
import plotly.express as px

gdf=df.groupby('sex').agg({'tip':'sum'})
display(gdf)
px.pie(gdf, values='tip', names=gdf.index, title='Sum of the tips')
```

## More Studies

### Percentage view

```python
gdf=df.groupby('sex').agg({'percentage':'mean'})
px.box(df, title = 'Precentage', y='percentage', x='sex')
```

### Who paid more tips?

```python
gdf=df.groupby('sex').agg({'tip':'sum'})
px.pie(gdf, values='tip', names=gdf.index, title='Sum of the tips')
```

### Best work day?

```python
gdf2=df.groupby('day').agg({'percentage':'mean'})
px.bar(gdf2, title = 'Precentage of the week days',text_auto=True).show()
```

## Seaice

```python
import pandas as pd
import plotly.express as px

seaice_data = pd.read_csv("seaice.csv", parse_dates=['Date'])

# Plot mit Plotly erstellen
fig = px.bar(seaice_data, x='Date', y='Extent',
              title='Arctic Ice Coverage')
fig.show()
```

```python
seaice_data.set_index('Date', inplace = True)
seaice_data_y = seaice_data.resample('YE').mean()
px.bar(seaice_data_y.reset_index(),                   
        title = 'Arctic Ice Coverage',
        x = 'Date',
        y = 'Extent',
        color = 'Extent',
        color_continuous_scale='Thermal_r'
       ).show()
```
