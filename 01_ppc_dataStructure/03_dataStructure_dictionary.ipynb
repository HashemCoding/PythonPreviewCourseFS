# Dictionary

## dict example

## create dict

```python
by_constructor = dict()
by_shorthand = {}
print(type(by_constructor), type(by_shorthand))
```

```python
my_dict = {'name': 'John', 'age': None, 'city': 'New York'}
print(my_dict)
```

## elements

```python
# Wert in eine Variabele speichern
age = my_dict['age']
print(age)
```

```python
# Hinzufgen eines neuen Schlssel-Wert-Paares
my_dict['occupation'] = 'Engineer'
print(my_dict)
```

```python
# Aktualisieren eines bestehenden Werts
my_dict['age'] = 26
print(my_dict)
```

```python
# Entfernen eines Schlssel-Wert-Paares
removed_value = my_dict.pop('city')
print(removed_value)
print(my_dict)
```

```python
# Alternativ: Lschen mit 'del'
del my_dict['name']
print(my_dict)
```

## Iterration

```python
# Iteration ber Schlssel
for key in my_dict:
    print(key)
```

```python
# Iteration ber Werte
for value in my_dict.values():
    print(value)
```

```python
# Iteration ber Schlssel-Wert-Paare
for key, value in my_dict.items():
    print(f"{key}: {value}")
```

## nested

```python
nested_dict = {'person': {'name': 'Alice', 'age': 30}}
print(nested_dict["person"])
print(nested_dict["person"]["name"])
removed= nested_dict["person"].popitem()
print(removed)
```

# Help

``len()`` - Gibt die Anzahl der Schlssel-Wert-Paare im Dictionary zurck.


``keys()`` - Gibt eine Liste der Schlssel im Dictionary zurck.


``values()`` - Gibt eine Liste der Werte im Dictionary zurck


``items()`` - Gibt eine Liste von Schlssel-Wert-Tupeln im Dictionary zurck

``get(key, default)`` - Gibt den Wert fr den angegebenen Schlssel zurck. Falls der Schlssel nicht existiert, wird default zurckgegeben)

``pop(key, default)`` - Entfernt den Schlssel und gibt den zugehrigen Wert zurck. Falls der Schlssel nicht existiert, wird default zurckgegeben)


``popitem()`` - Entfernt und gibt ein beliebiges Schlssel-Wert-Paar zurck

``update(dictionary)`` - Aktualisiert das Dictionary mit Schlssel-Wert-Paaren aus einem anderen Dictionaryn


``clear()`` - Entfernt alle Schlssel-Wert-Paare aus dem Dictionary


``copy()`` - Erzeugt eine Kopie des Dictionaries
