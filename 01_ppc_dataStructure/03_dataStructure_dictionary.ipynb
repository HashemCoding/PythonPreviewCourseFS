# Dictionary

## dict example

## create dict

```python
by_constructor = dict()
by_shorthand = {}
print(type(by_constructor), type(by_shorthand))
```

```python
my_dict = {'name': 'John', 'age': None, 'city': 'New York'}
print(my_dict)
```

## elements

```python
# Wert in eine Variabele speichern
age = my_dict['age']
print(age)
```

```python
# Hinzufügen eines neuen Schlüssel-Wert-Paares
my_dict['occupation'] = 'Engineer'
print(my_dict)
```

```python
# Aktualisieren eines bestehenden Werts
my_dict['age'] = 26
print(my_dict)
```

```python
# Entfernen eines Schlüssel-Wert-Paares
removed_value = my_dict.pop('city')
print(removed_value)
print(my_dict)
```

```python
# Alternativ: Löschen mit 'del'
del my_dict['name']
print(my_dict)
```

## Iterration

```python
# Iteration über Schlüssel
for key in my_dict:
    print(key)
```

```python
# Iteration über Werte
for value in my_dict.values():
    print(value)
```

```python
# Iteration über Schlüssel-Wert-Paare
for key, value in my_dict.items():
    print(f"{key}: {value}")
```

## nested

```python
nested_dict = {'person': {'name': 'Alice', 'age': 30}}
print(nested_dict["person"])
print(nested_dict["person"]["name"])
removed= nested_dict["person"].popitem()
print(removed)
```

# Help

``len()`` - Gibt die Anzahl der Schlüssel-Wert-Paare im Dictionary zurück.


``keys()`` - Gibt eine Liste der Schlüssel im Dictionary zurück.


``values()`` - Gibt eine Liste der Werte im Dictionary zurück


``items()`` - Gibt eine Liste von Schlüssel-Wert-Tupeln im Dictionary zurück

``get(key, default)`` - Gibt den Wert für den angegebenen Schlüssel zurück. Falls der Schlüssel nicht existiert, wird default zurückgegeben)

``pop(key, default)`` - Entfernt den Schlüssel und gibt den zugehörigen Wert zurück. Falls der Schlüssel nicht existiert, wird default zurückgegeben)


``popitem()`` - Entfernt und gibt ein beliebiges Schlüssel-Wert-Paar zurück

``update(dictionary)`` - Aktualisiert das Dictionary mit Schlüssel-Wert-Paaren aus einem anderen Dictionaryn


``clear()`` - Entfernt alle Schlüssel-Wert-Paare aus dem Dictionary


``copy()`` - Erzeugt eine Kopie des Dictionaries
